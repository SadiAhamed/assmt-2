## 3. Plot a bar graph of male vs female buyers participated in the sales.

```python
male=print("the number of males are",data["Buyer Gender"].value_counts()["Male"])
```

```python
female=print("the number of females are",data["Buyer Gender"].value_counts()["Female"])
```

```python
plt.figure()
plt.bar(data["Buyer Gender"],height=0.8)
plt.title("Male vs Female",fontsize=20)
plt.xlabel("gender")
plt.ylabel("freq")
plt.grid()
```

```python
sns.countplot(x="Buyer Gender",data=data,color="blue",saturation=0.1)
plt.title("Male vs Female",fontsize=20)
plt.xlabel("gender")
plt.grid()
plt.show()
```
